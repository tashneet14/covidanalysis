### using Data Analysis and Data Visualisation in Python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  #importing necessary libraries
```

```python
data=pd.read_csv(r"C:\Users\lenovo\Downloads\archive (1)\day_wise.csv")#importing data set
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Add x-axis and y-axis
ax.bar(a.index.values,
       a['Confirmed'],
       color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Confirmed cases",
       title="The surge of Covid-19")

plt.show()
```

## Heatmap

```python
sns.heatmap(data.corr()) #implementation of heatmap
```

***The above heatmap shows the positive and negative correlations of various columns wrt each other using mentioned color density***

## Relation plots

```python
sns.relplot(x='Confirmed',y='Active',data=data,kind='line')
```

***The above graph suggests that the active cases increased with the increase in confirmed cases but gradually flattens due to impose of lockdown and strict measures applied in all the countries***

```python
sns.relplot(x='Confirmed',y='Deaths',data=data,kind='scatter',hue='New deaths')
```

***According to the plot initially deaths increased at a higher rate wrt confirmed cases but after global 2,00,000 deaths starts reducing due to improved healthcare and knowledge related to covid with time***

```python
sns.relplot(x='Confirmed',y='Recovered',data=data,kind='scatter',hue='New recovered')
```

***Recovered cases have increased at a good pace in the mentioned time period depicting that covid has lost its grip with passing of time***

## Bar plot

```python
sns.barplot(x='Confirmed',y='No. of countries',data=data)
```

***The spread of corona virus was limited to nearly 25 countries for brief period of time but due to effect of continuous increase in confirmed cases and some countries reluctant to seal the borders and imposing strict lockdown rules more and more countries start getting affected and soon it was declared pandemic by WHO and within some days the deadly virus wrecked havoc on 90% of the globe.***

## Pair plot

```python
sns.pairplot(data)
```

***This is the pair plot depicting plotting of each column wrt all other columns***

## Distplot

```python
sns.distplot(data['Recovered / 100 Cases'],color='red')
```

***The density is highest for the range 50-60 which means for most of the days 50-60 patients were recovered per 100 cases***

## Jointplot

```python
sns.jointplot(x='New cases',y='New deaths',data=data,kind='hex')
```

***The plot suggests that after the initial surge, most new deaths(4000-8000) were reported when the new cases were ranging from 60,000 to 1,40,000 approximately***

## MATPLOTLIB PLOT

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Add x-axis and y-axis
ax.plot(a.index.values,
       a['New recovered'],
       color='green')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="New recovered cases",
       title="Beating coronavirus")

plt.show()
```

***The increase in new recovered cases suggest that the virus gradually lost control during the specified time period***
